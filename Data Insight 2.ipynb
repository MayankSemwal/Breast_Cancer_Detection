```python
import pandas as pd
import numpy as np
from matplotlib.colors import ListedColormap
pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 1000)
```

```python
Features = pd.read_csv('Features.txt',sep="\t")
Samples=pd.read_csv('sample.txt',sep="\t",header=None)
LabelsData = pd.read_csv('Labels.txt',sep="\t",header=None)
Features.columns=Features.columns.str.replace('Entrez_Gene_Id','SAMPLE_ID')
```

```python
GeneTr= Features.T
Header = GeneTr.iloc[1]
GeneTr = GeneTr[2:]
GeneTr.columns = Header
GeneTr.reset_index(inplace=True,col_level=0)
GeneTr.rename(columns = {"SAMPLE_ID": "NEW NAME", 
                     "index":"SAMPLE_ID"}, 
                                 inplace = True) 
GeneTr.head()
```

```python
LabelsData=LabelsData.loc[0:]
LabelsData=LabelsData.reset_index(drop=True)
LabelsData.columns = LabelsData.iloc[0]
LabelsData=LabelsData.drop(LabelsData.index[0])
```

```python
Samples=Samples.loc[4:]
Samples=Samples.reset_index(drop=True)
Samples.columns = Samples.iloc[0]
Samples=Samples.drop(Samples.index[0])
Samples
```

```python
col_list=['SAMPLE_ID','PATIENT_ID']
Samples=Samples[col_list]
Samples
```

```python
dLabel = LabelsData.merge(Samples, on="PATIENT_ID", how = 'inner')
```

```python
dCombine=GeneTr.merge(dLabel, on="SAMPLE_ID", how = 'inner')
dCombine.to_csv(r'CombinedDataset.csv')
```
